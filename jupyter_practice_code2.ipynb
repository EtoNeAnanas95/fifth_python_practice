```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

valorant = pd.read_csv("valorant-stats_PALOMALO_SILNO.csv")
```

```python
valorant = valorant.sort_values("Price")

guns_prices = valorant["Price"]

guns_damage = valorant["HDMG_0"]

plt.plot(guns_prices, guns_damage, color="pink", marker="o", markersize=8, label="График соотношения цены и урона в тело")
plt.title("График соотношения цены и урона в тело")
plt.legend()
plt.xlabel("ЦЕНА")
plt.ylabel("УРОН")
plt.show()
```

```python
valorant = valorant.sort_values("BDMG_0")

guns_damage = valorant["BDMG_0"]

plt.plot(guns_damage, guns_prices, color="blue", marker="o", markersize=8, label="График соотношения цены и урона в голову")
plt.title("График соотношения цены и урона в голову")
plt.legend()
plt.xlabel("УРОН")
plt.ylabel("ЦЕНА")
plt.show()
```

```python
valorant = valorant.sort_values("LDMG_0")

guns_damage = valorant["LDMG_0"]

plt.plot(guns_prices, guns_damage, color="blue", marker="o", markersize=8, label="График соотношения цены и урона в ноги")
plt.title("График соотношения цены и урона в голову")
plt.legend()
plt.xlabel("ЦЕНА")
plt.ylabel("УРОН")
plt.show()
```
